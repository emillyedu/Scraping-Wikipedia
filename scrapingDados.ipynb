Deputado Ernani Satyro

```python
#tabela lateral

from bs4 import BeautifulSoup
import requests
import pandas as pd

url = 'https://pt.m.wikipedia.org/wiki/Ern%C3%A2ni_S%C3%A1tiro'
classTable = 'infobox infobox_v2'

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

info = soup.find('table', {'class': classTable})
df = pd.read_html(str(info))
df = pd.DataFrame(df[0])
df
```

```python
#corpo da página

from bs4 import BeautifulSoup
import requests
import pandas as pd

url = 'https://pt.m.wikipedia.org/wiki/Ern%C3%A2ni_S%C3%A1tiro'
classSection1 = 'mf-section-0'
classSection2 = 'mf-section-1'
classSection3 = 'mf-section-2'
classSection4 = 'mf-section-3'
classSection5 = 'mf-section-4'
classTable = 'toccolours'

res = requests.get(url)
res.encoding = "utf-8"
soup = BeautifulSoup(res.text, 'html.parser')

posts = soup.find_all(class_ = classSection1)
for post in posts:
    introduction = post.find_all('p')
    for p in introduction:
        print(p.text)

posts2 = soup.find_all(class_ = classSection2)
for post2 in posts2:
    text1 = post2.find_all('p')
    text2 = post2.find_all('ul')
    for p in text1:
        print(p.text)
    for ul in text2:
        print(ul.text)


posts2 = soup.find_all(class_ = classSection3)
for post2 in posts2:
    text1 = post2.find_all('p')
    text2 = post2.find_all('ul')
    for p in text1:
        print(p.text)
    for ul in text2:
        print(ul.text)


posts2 = soup.find_all(class_ = classSection4)
for post2 in posts2:
    text1 = post2.find_all('p')
    text2 = post2.find_all('ul')
    for p in text1:
        print(p.text)
    for ul in text2:
        print(ul.text)


posts2 = soup.find_all(class_ = classSection5)
for post2 in posts2:
    text1 = post2.find_all('p')
    text2 = post2.find_all('ul')
    for p in text1:
        print(p.text)
    for ul in text2:
        print(ul.text)

info = soup.find('table', {'class': classTable})
df = pd.read_html(str(info))
df = pd.DataFrame(df[0])
df
```

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

wiki_url = 'https://pt.m.wikipedia.org/wiki/Lista_de_parlamentares_da_Para%C3%ADba'
table = 'wikitable'

response = requests.get(wiki_url)
soup = BeautifulSoup(response.text, 'html.parser')

total_table = soup.find_all('table', {'class':table})[1]
df = pd.read_html(str(total_table))
df = pd.DataFrame(df[0])
df
```
